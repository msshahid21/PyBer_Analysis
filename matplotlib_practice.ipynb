```python
# Import Dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() - Second way of doing the same as above
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() - Third Way, only two lines, no figure
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
#plt.show()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = "Boston")

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Create the plot
fig, ax = plt.subplots()

# Add the x and y axes with label
ax.plot(x_axis, y_axis, label = "Boston", color = "green", marker = "D", linewidth = 2)

# Create labels for the x and y axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")

# Set the y limit between 0 and 45
ax.set_ylim(0, 45)

# Create a title
ax.set_title("PyBer Fare by Month")

# Add a grid
ax.grid()

# Add the legend
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot
plt.barh(x_axis, y_axis)

# Invert the y_axis
plt.gca().invert_yaxis()
```

```python
# Skill Drill
# Create the plot
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")

# Invert the y_axis
plt.gca().invert_yaxis()

# Axes Labels
plt.xlabel("Fare ($)")
plt.ylabel("Date")

# Title
plt.title("PyBer Fare By Month")

# Legend
plt.legend()

```
